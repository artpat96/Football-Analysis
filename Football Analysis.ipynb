```python
#Function to obtain wanted data on the plot
def vis(data,players = Players):
    for i in players:
        plt.plot(Seasons, data[Pdict[i]], c=Col[i], ls='--', marker=Mar[i], ms=8, label=i)
    plt.xlabel('Season')
    plt.legend(loc=0, bbox_to_anchor=(1,1))
    plt.show()
```
